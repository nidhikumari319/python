# Q3.Compare and contrast mutable and immutable objects in Python with examples? ans:

Mutable Objects: These are objects whose value can be changed after they are created.

Examples : lists, dictionaries, and sets.

Immutable Objects: These are objects whose value cannot be changed after they are created.

Examples : integers, floats, strings, and tuples.

Comparison:

Can You Change It?

Mutable: Yes, you can change what’s inside. For example, you can add or remove items.

Immutable: No, you can’t change what’s inside. If you want to make a change, you have to create a new one.

Does It Stay the Same?

Mutable: The object stays the same in memory, even if you change its content.

Immutable: Changing it means making a new object. The old one stays the same.
